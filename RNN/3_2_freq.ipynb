```python
import operator

import nltk
import collections

nltk.download('webtext')
# print(nltk.corpus.webtext.fileids()
```

퀴즈
webtext 코퍼스에 있는 wine.txt 파일을 토큰으로 만드세요

```python

wine = nltk.corpus.webtext.raw('wine.txt')
wine = wine.lower()     # 소문자로 변경
# print(wine)

tokens = nltk.tokenize.RegexpTokenizer(r'\w+').tokenize(wine)
print(tokens[:10])

print(nltk.corpus.stopwords.fileids())

print('-'*100)

# stop_words : 불용어 목록
stop_words = nltk.corpus.stopwords.words('english')
print(stop_words)
```

### 퀴즈
### 토큰 목록으로부터 불용어와 1글자 토큰을 제거하세요

```python
tokens = [t for t in tokens if t not in stop_words]
print('불용어를 제거하고 난 수:', len(tokens))

# 한글자 지우기
tokens = [t for t in tokens if len(t) > 1]
print('불용어를 제거한 다음 한 글자 토큰까지 지우고 난 수: ', len(tokens))
```

### 퀴즈
### 단어별 빈도를 딕셔너리에 저장하세요

```python
freq = {}
for t in tokens:
    if t in freq:
        freq[t] += 1
    else:
        freq[t] = 1
print(freq)
```

```python
freq_1 = {}
for t in tokens:
    freq[t] = freq.get(t, 0) +1
print(freq_1)
```

```python
freq = collections.defaultdict(int)
for t in tokens:
    freq[t] += 1
print(freq)
```

### 퀴즈
### 빈도에 따라 정렬하세요(리스트 생성해야 한다.)

```python
# sorted(): 반복 가능한 객체(리스트, 튜플 등)를 받아들이고 해당 객체의 요소들을 정렬한 새로운 리스트 반환
freq_sorted = sorted(freq.items(), key=lambda t: t[1], reverse=True)     # t : t ->   , reverse = True : 내림차순
print(freq_sorted[:5])
```

```python
import operator

freq_sorted_1 = sorted(freq.items(), key=operator.itemgetter(1), reverse=True)
print(freq_sorted_1[:5])
```

```python
vocab = [t[0] for t in freq_sorted[:2000]]
print(vocab[:10])
```

```python
freq = nltk.FreqDist(tokens)        # 분포에 대해서 계산
print(freq)
print(freq.most_common(10))
```

## 퀴즈
### 1. json을 사용해서 예쁘게 출력
### 2. 정규 표현식을  사용해서 예쁘게 출력

```python
import json

for m in movies:
    item = json.loads(m)
    rank = item['currentRank']
    node = item['node']
    title = node['titleText']['text']
    year = node ['releaseYear']['year']
    runtime = node['runtime']['seconds']
    ratings = node['ratingsSummary']['aggregateRating']
    # print(node.keys())

    print(title, year, runtime, ratings)
    print('등수:', rank)
    print('제목:', title)
    print('연도:', year)
    print('시간:', '{}시간 {}분'.format(int(runtime) // 3600, int(runtime) % 3600 // 60))
    print('평점:', ratings)
```
